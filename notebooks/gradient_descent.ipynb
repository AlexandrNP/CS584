```python
from sklearn.metrics import mean_squared_error, r2_score
w = np.random.random(3)
print(w)

y_hat = np.matmul(X,w)
print(X*w)
print(y_hat)

mse = mean_squared_error(y, y_hat)
print(mse)
print(r2_score(y, y_hat))
```
